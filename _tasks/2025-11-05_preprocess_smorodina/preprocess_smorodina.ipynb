```python
import pandas as pd
import sys
import re
```

```python
df = pd.read_csv('smorodina_data.csv')
df = df.rename(columns={'название': 'Название',
                        'цена': 'Цена',
                        'применение': 'Инструкция по применению',
                        'активные ингридиенты': 'Состав',
                        'ссылка': 'Ссылка',
                        'описание' : 'Описание'})
```

```python
def get_price_number(string):
    match = re.search(r'\d+(\.\d+)?', string)
    if match:
        return match.group()
    else:
        return 300
```

```python
df['Цена'] = df['Цена'].apply(get_price_number).astype('int')
```

```python
df.to_csv('smorodina_data.csv', index=False)
```
